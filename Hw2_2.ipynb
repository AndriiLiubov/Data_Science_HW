- Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

- Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
import pandas as pd

developers = pd.read_csv("2017_jun_final.csv")


python_data = developers[developers["Мова програмування"] == "Python"]

grouped_by_position = python_data.groupby("Посада")

def fill_avg_salary(group):
    group["avg"] = group["Зарплата на місяць"].mean()
    return group

python_data = grouped_by_position.apply(fill_avg_salary)


python_data
```

- Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
avg_stats = python_data["avg"].describe()


avg_stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Загальний досвід роботи', y='Зарплата на місяць', data=developers, s=100, edgecolor='w', alpha=0.7)
plt.title('Зависимость зарплаты от общего опыта работы', fontsize=16)
plt.xlabel('Общий опыт работы (годы)', fontsize=14)
plt.ylabel('Зарплата на месяц', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Загальний досвід роботи', y='Зарплата на місяць', data=developers, marker='o')
plt.title('Изменение зарплаты в зависимости от общего опыта работы', fontsize=16)
plt.xlabel('Общий опыт работы (годы)', fontsize=14)
plt.ylabel('Зарплата на месяц', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Рівень англійської', y='Зарплата на місяць', data=developers)
plt.title('Распределение зарплаты по уровням английского', fontsize=16)
plt.xlabel('Уровень английского', fontsize=14)
plt.ylabel('Зарплата на месяц', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
