- **Пошук та сортування даних**

```python
highest_rating = df['user_rating'].max()
print(f"Відповідь: Який рейтинг у датасеті найвищий? Відповідь: {highest_rating}")

highest_rating_count = df[df['user_rating'] == highest_rating].shape[0]
print(f"Відповідь: Скільки книг мають такий рейтинг? Відповідь: {highest_rating_count}")

most_reviews_book = df[df['reviews'] == df['reviews'].max()]['name'].values[0]
print(f"Відповідь: Яка книга має найбільше відгуків? Відповідь: {most_reviews_book}")

top_50_2015 = df[df['year'] == 2015]
most_expensive_book_2015 = top_50_2015[top_50_2015['price'] == top_50_2015['price'].max()]['name'].values[0]
print(f"Відповідь: З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча? Відповідь: {most_expensive_book_2015}")

fiction_2010_count = df[(df['genre'] == 'Fiction') & (df['year'] == 2010)].shape[0]
print(f"Відповідь: Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році? Відповідь: {fiction_2010_count}")

rating_49_2010_2011_count = df[(df['user_rating'] == 4.9) & (df['year'].isin([2010, 2011]))].shape[0]
print(f"Відповідь: Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках? Відповідь: {rating_49_2010_2011_count}")

sorted_books_2015 = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values(by='price')
last_book_sorted = sorted_books_2015['name'].values[-1]
print(f"Відповідь: Яка книга остання у відсортованому списку? Відповідь: {last_book_sorted}")
```

- Для початку давайте подивимося на максимальну та мінімальну ціни для кожного з жанрів (використовуйте функції groupby та agg, для підрахунку мінімальних та максимальних значень використовуйте max та min). Не беріть усі стовпці, виберіть тільки потрібні вам

Відповідь: Максимальна ціна для жанру Fiction: Відповідь

Відповідь: Мінімальна ціна для жанру Fiction: Відповідь

Відповідь: Максимальна ціна для жанру Non Fiction: Відповідь

Відповідь: Мінімальна ціна для жанру Non Fiction: Відповідь

```python
genre_price_stats = df.groupby('genre')['price'].agg(['max', 'min']).reset_index()
print(f"Відповідь: Максимальна ціна для жанру Fiction: {genre_price_stats[genre_price_stats['genre'] == 'Fiction']['max'].values[0]}")
print(f"Відповідь: Мінімальна ціна для жанру Fiction: {genre_price_stats[genre_price_stats['genre'] == 'Fiction']['min'].values[0]}")
print(f"Відповідь: Максимальна ціна для жанру Non Fiction: {genre_price_stats[genre_price_stats['genre'] == 'Non Fiction']['max'].values[0]}")
print(f"Відповідь: Мінімальна ціна для жанру Non Fiction: {genre_price_stats[genre_price_stats['genre'] == 'Non Fiction']['min'].values[0]}")
```

- Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

Відповідь: Якої розмірності вийшла таблиця? Відповідь:

Відповідь: Який автор має найбільше книг? Відповідь:

Відповідь: Скільки книг цього автора? Відповідь:

```python
author_book_count = df.groupby('author')['name'].count().reset_index()
author_book_count.columns = ['author', 'book_count']
print(f"Відповідь: Якої розмірності вийшла таблиця? Відповідь: {author_book_count.shape}")
most_books_author = author_book_count[author_book_count['book_count'] == author_book_count['book_count'].max()]
print(f"Відповідь: Який автор має найбільше книг? Відповідь: {most_books_author['author'].values[0]}")
print(f"Відповідь: Скільки книг цього автора? Відповідь: {most_books_author['book_count'].values[0]}")
```

- Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

Відповідь: У якого автора середній рейтинг мінімальний? Відповідь:

Відповідь: Який у цього автора середній рейтинг? Відповідь:

```python
author_avg_rating = df.groupby('author')['user_rating'].mean().reset_index()
author_avg_rating.columns = ['author', 'avg_rating']
min_avg_rating_author = author_avg_rating[author_avg_rating['avg_rating'] == author_avg_rating['avg_rating'].min()]
print(f"Відповідь: У якого автора середній рейтинг мінімальний? Відповідь: {min_avg_rating_author['author'].values[0]}")
print(f"Відповідь: Який у цього автора середній рейтинг? Відповідь: {min_avg_rating_author['avg_rating'].values[0]:.2f}")
```

- З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

```python
combined_df = pd.concat([author_book_count.set_index('author'), author_avg_rating.set_index('author')], axis=1).reset_index()

combined_df
```

- Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)

Відповідь: Який автор перший у списку?

```python
sorted_combined_df = combined_df.sort_values(by=['book_count', 'avg_rating']).reset_index(drop=True)
print(f"Відповідь: Який автор перший у списку? Відповідь: {sorted_combined_df['author'].values[0]}")
```
